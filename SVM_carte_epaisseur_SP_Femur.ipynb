```python
import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt

Categories=['Femur Patho','Femur Sain']
flat_data_arr_train=[] #input array
target_arr_train=[] #output array
datadir='/content/drive/MyDrive/cartes_epaisseur' 
#path which contains all the categories of images

list_train = []
list_test = []
train_ids_fem_pato = [1,6,7,12,13,19,25]
test_ids_fem_pato = [20,22]
train_ids_fem_sain = [18,65,66,67,68,69]
test_ids_fem_sain = [71,74,60,62,64]

for i in train_ids_fem_pato:
    image_name = '/femur_patho/Femur_%02d.png'%(i)
    list_train.append(datadir + image_name)

for i in test_ids_fem_pato:
    image_name = '/femur_patho/Femur_%02d.png'%(i)
    list_test.append(datadir + image_name)

for i in train_ids_fem_sain:
    image_name = '/femur/Femur_%02d.png'%(i)
    list_train.append(datadir + image_name)

for i in test_ids_fem_sain:
    image_name = '/femur/Femur_%02d.png'%(i)
    list_test.append(datadir + image_name)



i=0
for img in list_train:
      img_array=imread(img)
      flat_data_arr_train.append(img_array.flatten())
      if i<len(train_ids_fem_pato):
        target_arr_train.append(Categories.index('Femur Patho'))
      else:
        target_arr_train.append(Categories.index('Femur Sain'))
      i+=1

flat_data_train=np.array(flat_data_arr_train)
target_train=np.array(target_arr_train)
df_train=pd.DataFrame(flat_data_train) #dataframe
df_train['Target']=target_train
x_train=df_train.iloc[:,:-1] #input data 
y_train=df_train.iloc[:,-1] #output data
```

```python
flat_data_arr_test=[] #input array
target_arr_test=[] #output array

j=0
for img in list_test:
      img_array=imread(img)
      flat_data_arr_test.append(img_array.flatten())
      if j<len(test_ids_fem_pato):
        target_arr_test.append(Categories.index('Femur Patho'))
      else:
        target_arr_test.append(Categories.index('Femur Sain'))
      j+=1

flat_data_test=np.array(flat_data_arr_test)
target_test=np.array(target_arr_test)
df_test=pd.DataFrame(flat_data_test) #dataframe
df_test['Target']=target_test
x_test=df_test.iloc[:,:-1] #input data 
y_test=df_test.iloc[:,-1] #output data
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly']}
svc=svm.SVC(probability=True)

svc.fit(x_train,y_train)
print('The Model is trained well with the given images')
model=GridSearchCV(svc,param_grid)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

y_pred=model.predict(x_test)
print("The predicted Data is :")
print(y_pred)
print("The actual data is:")
print(np.array(y_test))
print(f"The model is {accuracy_score(y_test,y_pred)*100}% accurate")
print("un petit text ", model.best_params_)
print("f1:")
print(f1_score(y_test,y_pred, average=None))
print("precision:")
print(precision_score(y_test,y_pred, average=None))
print("recall:")
print(recall_score(y_test,y_pred, average=None))
```
